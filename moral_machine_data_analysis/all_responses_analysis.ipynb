```python
import pandas as pd

chunk_size = 100000  # Number of rows per chunk
file_path = '/Users/buseaydin/Desktop/492/moral machine/Datasets/Moral Machine Data/SharedResponses.csv'
#chunks = pd.read_csv(file_path, low_memory=False)
chunks = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)

# Example: Combine chunks into a DataFrame (optional)
df = pd.concat(chunks, ignore_index=True)
```

```python
pd.set_option('display.max_columns', None)
print(df.head(50))
```

```python
column_names = df.columns.tolist()
print("Column Names:", column_names)

# Get the total number of rows
total_rows = len(df)
print("Total Number of Rows:", total_rows)
```

```python
sorted_df = df.sort_values(by='ResponseID')
```

```python
sorted_df
```

```python
# Save to Parquet
sorted_df['Man'] = pd.to_numeric(sorted_df['Man'], errors='coerce')

sorted_df.to_parquet('sorted_filtered_df.parquet')
```

```python
# Count occurrences of each 'ResponseID'
response_counts = sorted_df['ResponseID'].value_counts()

# Filter 'ResponseID' values with exactly 2 occurrences
valid_responses = response_counts[response_counts == 2].index

# Filter the DataFrame to keep only rows with 'ResponseID' that have exactly 2 occurrences
sorted_filtered_groups = sorted_df[sorted_df['ResponseID'].isin(valid_responses)]
```
