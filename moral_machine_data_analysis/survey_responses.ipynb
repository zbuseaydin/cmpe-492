```python
import pandas as pd

chunk_size = 100000  # Number of rows per chunk
pd.set_option('display.max_columns', None)
```

```python
survey_file_path = '/Users/buseaydin/Desktop/492/moral machine/Datasets/Moral Machine Data/SharedResponsesSurvey.csv'
survey_chunks = pd.read_csv(survey_file_path, chunksize=chunk_size, low_memory=False)

survey_df = pd.concat(survey_chunks, ignore_index=True)
```

```python
filtered_all_file_path = '/Users/buseaydin/Desktop/492/moral machine/data_analysis/moral_machine/attributes_filtered.csv'
filtered_all_chunks = pd.read_csv(filtered_all_file_path, chunksize=chunk_size, low_memory=False)

filtered_all_df = pd.concat(filtered_all_chunks, ignore_index=True)
```

```python
merged_df = pd.merge(survey_df, filtered_all_df, on=['ResponseID', 'Saved'], how='inner')
```

```python
merged_df = merged_df.sort_values(by='ResponseID')
merged_df.reset_index(drop=True, inplace=True)
merged_df
```

```python
saved_responses = merged_df[merged_df['Saved'] == 1]
saved_responses
```

```python
unique_review_age = merged_df['Review_age'].unique()
unique_review_education = merged_df['Review_education'].unique()
unique_review_gender = merged_df['Review_gender'].unique()
unique_review_income = merged_df['Review_income'].unique()
unique_review_political = merged_df['Review_political'].unique()
unique_review_religious = merged_df['Review_religious'].unique()

# Print the unique values
print("Unique Review_age values:", unique_review_age)
print("Unique Review_education values:", unique_review_education)
print("Unique Review_gender values:", unique_review_gender)
print("Unique Review_income values:", unique_review_income)
print("Unique Review_political values:", unique_review_political)
print("Unique Review_religious values:", unique_review_religious)
```

```python
# Ensure that 'Review_age' is numeric, then filter the rows based on the specified conditions
filtered_df = saved_responses[
    (saved_responses['Review_age'].apply(pd.to_numeric, errors='coerce') >= 0) & 
    (saved_responses['Review_age'].apply(pd.to_numeric, errors='coerce') <= 100) &  # Age should be between 0 and 100
    (saved_responses['Review_gender'].isin(['female', 'male', 'others'])) &  # Gender can only be female, male, or others
    (saved_responses['Review_religious'] >= 0) & (saved_responses['Review_religious'] <= 1) &  # Religious should be between 0 and 1
    (saved_responses['Review_political'] >= 0) & (saved_responses['Review_political'] <= 1)  # Political should be between 0 and 1
]

# Optionally, reset the index after filtering
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

```python
filtered_df.loc[:, 'Review_age'] = pd.to_numeric(filtered_df['Review_age'], errors='coerce')

filtered_unique_review_age = filtered_df['Review_age'].unique()
filtered_unique_review_education = filtered_df['Review_education'].unique()
filtered_unique_review_gender = filtered_df['Review_gender'].unique()
filtered_unique_review_income = filtered_df['Review_income'].unique()
filtered_unique_review_political = filtered_df['Review_political'].unique()
filtered_unique_review_religious = filtered_df['Review_religious'].unique()

# Print the unique values
print("Unique Review_age values:", filtered_unique_review_age)
print("Unique Review_education values:", filtered_unique_review_education)
print("Unique Review_gender values:", filtered_unique_review_gender)
print("Unique Review_income values:", filtered_unique_review_income)
print("Unique Review_political values:", filtered_unique_review_political)
print("Unique Review_religious values:", filtered_unique_review_religious)
```

```python
# filter for education
filtered_df = filtered_df[~filtered_df['Review_education'].isin(['default', 'others'])]

education_mapping = {
    'underHigh': 1,
    'high': 2,
    'vocational': 3,
    'college': 4,
    'bachelor': 5,
    'graduate': 6
}

filtered_df['Review_education_numeric'] = filtered_df['Review_education'].map(education_mapping)
```

```python
# filter for income
filtered_df = filtered_df[~filtered_df['Review_income'].isin(['default', 'over10000'])]

# Normalize and map income values
income_mapping = {
    'under5000': 1,
    '5000': 2,
    '10000': 3,
    '15000': 4,
    '25000': 5,
    '35000': 6,
    '50000': 7,
    '80000': 8,
    'above100000': 9
}
filtered_df['Review_income_numeric'] = filtered_df['Review_income'].map(income_mapping)
```

```python
# filter for gender
filtered_df = filtered_df[~filtered_df['Review_gender'].isin(['others'])]

# Map gender values to numeric
gender_mapping = {
    'female': 1,
    'male': 2,
}
filtered_df['Review_gender_numeric'] = filtered_df['Review_gender'].map(gender_mapping)
```

```python
filtered_df
```

```python
scenario_types = ['Age', 'Gender', 'Fitness', 'Utilitarian', 'Social Status', 'Species']

# Create an empty dictionary to store the DataFrames
scenario_dfs = {}

# Loop through each unique ScenarioType_x and create a DataFrame for each
for scenario in scenario_types:
    scenario_dfs[scenario] = filtered_df[filtered_df['ScenarioType_x'] == scenario]

# Example: Access the DataFrame for 'Age' scenario
age_df = scenario_dfs['Age']
gender_df = scenario_dfs['Gender']
fitness_df = scenario_dfs['Fitness']
utilitarian_df = scenario_dfs['Utilitarian']
social_status_df = scenario_dfs['Social Status']
species_df = scenario_dfs['Species']
```

```python
age_df = age_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
gender_df = gender_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
fitness_df = fitness_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
utilitarian_df = utilitarian_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
social_status_df = social_status_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
species_df = species_df.drop(columns=['ExtendedSessionID', 'ScenarioOrder'])
```

```python
# List of scenario DataFrames and their names
scenario_dfs = [age_df, gender_df, fitness_df, utilitarian_df, social_status_df, species_df]
scenario_names = ["Age", "Gender", "Fitness", "Utilitarian", "Social Status", "Species"]
```

```python
# List of columns for which to compute averages
columns_to_average = ['Review_gender_numeric', 'Review_education_numeric', 'Review_income_numeric', 'Review_age', 'Review_political', 'Review_religious']

# Iterate over each scenario DataFrame
for scenario_name, df in zip(scenario_names, scenario_dfs):
    print(f"\n--- {scenario_name} Scenario ---")
    
    # Group by AttributeLevel_x
    attributes_grouped = df.groupby('AttributeLevel_x')
    
    # Calculate the percentages for each AttributeLevel_x
    percentages = df['AttributeLevel_x'].value_counts(normalize=True) * 100
    print(f"Percentage of responses for each AttributeLevel:\n{percentages}\n")
    
    # Compute averages for each predefined column and print
    for column in columns_to_average:
        if column in df.columns:  # Check if column exists in the current DataFrame
            average = attributes_grouped[column].mean()
            print(f"Average of {column}:\n{average}\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def radar_chart(averages, columns, title):
    # Number of variables (columns to average)
    num_vars = len(columns)

    # Compute the angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    closed_circle = angles.copy()
    closed_circle.append(0.0)

    closed_averages = averages + [averages[0]]

    ax = plt.subplot(polar=True)

    plt.xticks(angles, columns_to_average)
    
#    ax.set_rlabel_position(0)
#    plt.yticks([20,40,60,80,100], ['20', '40', '60', '80', '100'])
#    plt.ylim(0,100)

    ax.plot(closed_circle, closed_averages)
    plt.show()


for scenario_name, df in zip(scenario_names, scenario_dfs):
    normalized_df = df.copy()
    for column in columns_to_average:
        normalized_df[column] = (normalized_df[column] - normalized_df[column].min()) / (normalized_df[column].max() - normalized_df[column].min())
        
    attributes_grouped = normalized_df.groupby('AttributeLevel_x')
    averages = []
    for column in columns_to_average:
        averages.append(attributes_grouped[column].mean())
        
    print(scenario_name.capitalize())
    radar_chart(averages, columns_to_average, "scenario trial")
```
