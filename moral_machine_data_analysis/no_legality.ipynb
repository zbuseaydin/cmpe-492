```python
import pandas as pd

chunk_size = 100000  # Number of rows per chunk
file_path = 'attributes_filtered.csv'
chunks = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)

# Example: Combine chunks into a DataFrame (optional)
df = pd.concat(chunks, ignore_index=True)
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
no_legality_rows = df[df['CrossingSignal'] == 0]
```

```python
no_legality_rows
```

```python
ped_rows = no_legality_rows[no_legality_rows['PedPed'] == 1]
ped_rows
```

```python
barrier_rows = ped_rows[ped_rows['Barrier'] == 0]
barrier_rows
```

```python
non_random_rows = barrier_rows[barrier_rows['ScenarioType'] != 'Random']
non_random_rows
```

```python
non_random_rows.to_csv('all_filtered.csv')
```
