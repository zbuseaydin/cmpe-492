```python
import pandas as pd

chunk_size = 100000  # Number of rows per chunk
file_path = 'sorted_filtered_df.csv'
chunks = pd.read_csv(file_path, chunksize=chunk_size, low_memory=False)

# Example: Combine chunks into a DataFrame (optional)
df = pd.concat(chunks, ignore_index=True)
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
def filter_same_pedped_barrier(group):
    return group['PedPed'].nunique() == 1 and group['Barrier'].nunique() == 1

# Group by 'ResponseID' and filter based on the condition
filtered_df = df.groupby('ResponseID').filter(filter_same_pedped_barrier)

filtered_df
```

```python
crossing_signal_filtered_df = filtered_df.groupby('ResponseID').filter(lambda x: x['CrossingSignal'].nunique() == 1)

crossing_signal_filtered_df
```

```python
pd.options.display.max_rows = 200
```

```python
crossing_signal_filtered_df
```

```python
crossing_signal_filtered_df.to_csv('attributes_filtered.csv')
```

```python
# Filter rows where Saved = 1
chosen_rows = crossing_signal_filtered_df[crossing_signal_filtered_df['Saved'] == 1]

# Group by 'ScenarioType' and 'AttributeLevel', then count occurrences
chosen_counts = chosen_rows.groupby(['ScenarioType', 'AttributeLevel']).size().reset_index(name='Count')

# Calculate percentages
total_counts = chosen_counts.groupby('ScenarioType')['Count'].transform('sum')
chosen_counts['Percentage'] = (chosen_counts['Count'] / total_counts) * 100

# Display the result
print(chosen_counts)
```

```python
# Filter rows where Saved = 1
chosen_rows_intervention = crossing_signal_filtered_df[crossing_signal_filtered_df['Saved'] == 1]

# Group by 'ScenarioType', 'Intervention', and 'AttributeLevel' to count occurrences
chosen_counts_intervention = chosen_rows_intervention.groupby(['ScenarioType', 'Intervention', 'AttributeLevel']).size().reset_index(name='Count')

# Calculate the total counts for each 'ScenarioType'
scenario_totals_intervention = chosen_counts_intervention.groupby('ScenarioType')['Count'].transform('sum')

# Calculate percentages relative to the total count for each 'ScenarioType'
chosen_counts_intervention['Percentage'] = (chosen_counts_intervention['Count'] / scenario_totals_intervention) * 100

# Display the result
print(chosen_counts_intervention)
```

- Intervention: represents the decision of the AV (STAY or SWERVE) that would lead to this outcome [0: the character would die if the AV stays, 1: the character would die if AV swerves]. This is not the actual decision taken by the user, but rather a part of the structural characterisation of the scenario.
- Saved: this resembles the actual decision made by the user [1: user decided to save the characters in this outcome, 0: user decided to kill the characters in this outcome].
